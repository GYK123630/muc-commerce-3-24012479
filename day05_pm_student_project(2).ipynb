## 任务2：公共基础指标（必做）

```python
# TODO：构建overall_metrics DataFrame，至少包含以下指标：
# 用户数、流失人数、流失率、平均订单数、订单数中位数、
# 平均优惠券数、平均返现、平均App时长、平均满意度、平均距上次下单天数

overall_metrics = None

# display(overall_metrics)
import pandas as pd

# ---------------------- 构建overall_metrics ----------------------
# 1. 基础统计
total_users = len(df)
churn_users = df["Churn"].sum()
overall_churn_rate = churn_users / total_users  # 总体流失率，用于下方校验

avg_order_count = df["OrderCount"].mean()
median_order_count = df["OrderCount"].median()

avg_coupon = df["CouponUsed"].mean()
avg_cashback = df["CashbackAmount"].mean()
avg_app_time = df["HourSpendOnApp"].mean()
avg_satisfaction = df["SatisfactionScore"].mean()
avg_days_last_order = df["DaySinceLastOrder"].mean()

# 组装指标字典，转为DataFrame
metrics_dict = {
    "用户数": total_users,
    "流失人数": churn_users,
    "流失率": overall_churn_rate,
    "平均订单数": avg_order_count,
    "订单数中位数": median_order_count,
    "平均优惠券使用数": avg_coupon,
    "平均返现金额": avg_cashback,
    "平均App使用时长": avg_app_time,
    "平均满意度评分": avg_satisfaction,
    "平均距上次下单天数": avg_days_last_order
}
# 转为单行DataFrame
overall_metrics = pd.DataFrame([metrics_dict])

# 展示指标
display(overall_metrics)

# ---------------------- 检查点2（无需修改） ----------------------
assert isinstance(overall_metrics, pd.DataFrame), "overall_metrics应为DataFrame"
assert len(overall_metrics.columns) >= 10, "公共指标至少10项"
assert abs(overall_churn_rate - 0.16838365896980462) < 1e-8, "总体流失率不正确"
print("检查点2通过")
```

## 任务3：单维专题分析（必做）

请选择一个专题：

- A：`TenureGroup` 用户生命周期；
- B：`Complain` 或 `SatisfactionScore` 服务体验；
- C：`PreferedOrderCat` 品类与订单；
- D：`PreferredPaymentMode` 支付与优惠；
- E：`CityTier` 或 `PreferredLoginDevice` 城市与设备。

最低要求：使用 `groupby + agg`，同时输出用户数和至少3项业务指标。

```python
# TODO: 填写你的分组字段
segment_field = "CityTier"

# TODO: 使用groupby + agg完成聚合，适配旧版pandas
segment_analysis = df.groupby(segment_field).agg({
    "CustomerID":"nunique",
    "OrderCount":"sum",
    "CashbackAmount":"mean",
    "SatisfactionScore":"mean"
}).rename(columns={
    "CustomerID":"用户数",
    "OrderCount":"订单总数",
    "CashbackAmount":"平均返现金额",
    "SatisfactionScore":"平均满意度"
}).reset_index()

# TODO: 重置索引、排序并展示
segment_analysis = segment_analysis.sort_values(by="用户数", ascending=False).reset_index(drop=True)
display(segment_analysis)

# 检查点（固定不动）
assert segment_field in df.columns, "segment_field不是有效字段"
assert isinstance(segment_analysis, pd.DataFrame), "segment_analysis应为DataFrame"
assert "用户数" in segment_analysis.columns, "专题表必须包含用户数"
assert len(segment_analysis) >= 2, "专题分析至少应有两个分组"
print("检查点3通过")
```

## 任务4：双维度交叉分析（必做）

```python
# TODO：从以下维度中选择两个
# TenureGroup、Complain、PreferedOrderCat、CityTier、PreferredLoginDevice
dim_1 = "CityTier"
dim_2 = "TenureGroup"

# TODO：按两个维度统计用户数、流失人数、流失率，以及至少一个行为指标
cross_analysis = df.groupby([dim_1, dim_2]).agg({
    "CustomerID":"nunique",
    "Churn":"sum",
    "OrderCount":"mean"
}).rename(columns={
    "CustomerID":"用户数",
    "Churn":"流失人数",
    "OrderCount":"平均订单数"
}).reset_index()

# 计算流失率
cross_analysis["流失率"] = cross_analysis["流失人数"] / cross_analysis["用户数"]

# TODO：新增“样本提示”列；用户数<30标记为“小样本”，否则为“可观察”
cross_analysis["样本提示"] = cross_analysis.apply(lambda row: "小样本" if row["用户数"] < 30 else "可观察", axis=1)

# TODO：按流失率或用户数排序并展示
cross_analysis = cross_analysis.sort_values(by="流失率", ascending=False).reset_index(drop=True)
display(cross_analysis)

# 检查点固定代码
assert dim_1 in df.columns and dim_2 in df.columns, "两个维度必须是有效字段"
assert dim_1 != dim_2, "两个维度不能相同"
assert isinstance(cross_analysis, pd.DataFrame), 'cross_analysis应为DataFrame'
assert {"用户数", "流失率", "样本提示"}.issubset(cross_analysis.columns), "双维表缺少必需列"
assert set(cross_analysis["样本提示"]).issubset({"小样本", "可观察"}), "样本提示取值不正确"
print("检查点4通过")
```

## 任务5：报表输出与回读验证（必做）

```python
# TODO：将三个表导出到OUTPUT_DIR
# 文件名必须为：overall_metrics.csv、segment_analysis.csv、cross_analysis.csv
# 要求：index=False，encoding="utf-8-sig"

outputs = {
    "overall_metrics.csv": overall_metrics,
    "segment_analysis.csv": segment_analysis,
    "cross_analysis.csv": cross_analysis,
}

# TODO：循环导出并重新读取；打印每个文件的shape
import os
import pandas as pd

# 定义输出文件夹（题目给定OUTPUT_DIR）
OUTPUT_DIR = "./output"
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

# 1. 将3个数据表放进字典，键为文件名，值为DataFrame
outputs = {
    "overall_metrics.csv": overall_metrics,
    "segment_analysis.csv": segment_analysis,
    "cross_analysis.csv": cross_analysis
}

# 2. 循环导出文件，index=False，编码utf‑8‑sig适配中文
for filename, table in outputs.items():
    path = os.path.join(OUTPUT_DIR, filename)
    table.to_csv(path, index=False, encoding="utf-8-sig")

# -------- 检查点5：重新读取文件并校验形状 --------
for filename, table in outputs.items():
    path = os.path.join(OUTPUT_DIR, filename)
    assert os.path.exists(path), f"缺少输出文件:{filename}"
    reloaded = pd.read_csv(path, encoding="utf-8-sig")
    assert reloaded.shape == table.shape, f"{filename}回读形状不一致"

print("检查点5通过：三个表均成功导出并回读。")
```
